## **Change birthdate to date**

```python
hr['birthdate'].head()
```

```python
def convert_date(date_str):

  if pd.isnull(date_str):
      return pd.NaT

  today = date.today()

  try:
      date_obj = pd.to_datetime(date_str, format='%m-%d-%y', errors ="raise").date()
  except ValueError:
      try:
          date_obj = pd.to_datetime(date_str, format='%m/%d/%y', errors = "raise").date()
      except ValueError:
          try:
            date_obj = pd.to_datetime(date_str, errors='raise').date()
          except ValueError:
              return pd.NaT  # Not a Time

  if date_obj.year > today.year:
    date_obj = date_obj.replace(year=date_obj.year - 100)

  return date_obj
```

```python
hr['birthdate'] = hr['birthdate'].apply(convert_date)
```

```python
hr['hire_date'] = hr['hire_date'].apply(convert_date)
```

## **Convert termdate to date**

```python
def convert_termdate(termdate):

  if pd.isnull(termdate):
      return pd.NaT

  if 'UTC' in termdate:
    termdate = re.sub(r' \d{2}:\d{2}:\d{2} UTC', '', termdate).strip()

  try:
    date_obj = pd.to_datetime(termdate, errors='raise').date()
  except ValueError:
    return pd.NaT
  return date_obj
```

```python
hr['termdate'] = hr['termdate'].apply(convert_termdate)
```

## **5. What is the average length of employment for employees who have been terminated?**

```python
hr['termdate'] = pd.to_datetime(hr['termdate'])
hr['hire_date'] = pd.to_datetime(hr['hire_date'])

terminated_employees = hr[hr['termdate'].notnull()]
```

```python
avg_length = round(((terminated_employees['termdate'] - terminated_employees['hire_date']).dt.days / 365).mean())
avg_length
```

## **8. Which department has the highest turnover rate?**

```python
department_turn_over_rate = {}
total_employees = hr['department'].value_counts()
for department, group in terminated_employees.groupby('department'):
  department_turn_over_rate[department] = (len(group) / total_employees[department])
  print(f'len : {len(group)}')
  print(f'total: {total_employees[department]}')

department_turn_over_rate = pd.Series(department_turn_over_rate)
department_turn_over_rate.sort_values(ascending = False)
```

## **10. How has the company's employee count changed over time based on hire and term dates?**

```python
net_change_percent = []
hires_by_year = hr.groupby(hr['hire_date'].dt.year)['hire_date'].count()
terminations_by_year = terminated_employees.groupby(terminated_employees['termdate'].dt.year)['termdate'].count()
for year in hires_by_year.index:
  if year in terminations_by_year.index:
    change = round(((hires_by_year[year] -terminations_by_year[year])/hires_by_year[year]) * 100, 2)
    net_change_percent.append(change)
  else:
    net_change_percent.append(0)
net_change_percent = pd.Series(net_change_percent, index = hires_by_year.index, name = 'percent')
net_change_percent
```
